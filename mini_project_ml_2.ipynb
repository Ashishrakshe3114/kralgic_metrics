```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/realistic_kraljic_dataset.csv')
df
```

```python
df.columns
```

```python
# remove the 'product id , product name , supplier region '

df = df.drop(df.columns[[0,1,2]],axis=1)
df
```

```python
df['Single_Source_Risk']
```

```python
# convert yes =1 and no = 0
df['Single_Source_Risk']=df['Single_Source_Risk'].map({'Yes':1,'No':0})
```

```python
df['Single_Source_Risk']
```

```python
df.info()
```

```python
df.columns
```

```python
# divide into two parts input and output

x=df[['Lead_Time_Days', 'Order_Volume_Units', 'Cost_per_Unit',
       'Supply_Risk_Score', 'Profit_Impact_Score', 'Environmental_Impact',
       'Single_Source_Risk']]

y=df['Kraljic_Category']
```

```python
# import all model

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# call the model

model_knn = KNeighborsClassifier()
model_LR = LogisticRegression()
model_svm = SVC()
model_nb = GaussianNB()
```

```python
# train test splits

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
# train the model

model_knn.fit(x_train,y_train)
model_LR.fit(x_train,y_train)
model_svm.fit(x_train,y_train)
model_nb.fit(x_train,y_train)
```

```python
# testing the model

y_pred_knn = model_knn.predict(x_test)
y_pred_LR = model_LR.predict(x_test)
y_pred_svm = model_svm.predict(x_test)
y_pred_nb = model_nb.predict(x_test)
```

```python
#evalution

print(classification_report(y_test,y_pred_knn))
```

```python
print(classification_report(y_test,y_pred_LR))
```

```python
print(classification_report(y_test,y_pred_svm))
```

```python
print(classification_report(y_test,y_pred_nb))
```
